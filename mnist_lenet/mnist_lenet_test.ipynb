```python
import time
import numpy as np
import tensorflow as tf
from tensorflow.examples.tutorials.mnist import input_data
import mnist_lenet_backward
import mnist_lenet_forward

def test(mnist):
    with tf.Graph().as_default() as g:
        x = tf.placeholder(tf.float32,[mnist_lenet_backward.BATCH_SIZE,
                                      mnist_lenet_forward.IMAGE_SIZE,
                                      mnist_lenet_forward.IMAGE_SIZE,
                                      mnist_lenet_forward.NUM_CHANNELS])
        y_ = tf.placeholder(tf.float32,[None,mnist_lenet_forward.OUTPUT_SIZE])
        y = mnist_lenet_forward.forward(x,False,None)
        
        ema = tf.train.ExponentialMovingAverage(mnist_lenet_backward.MOVING_AVERAGE_DECAY)
        ema_restore = ema.variables_to_restore()
        saver = tf.train.Saver(ema_restore)
        
        correct_prediction = tf.equal(tf.argmax(y,1),tf.argmax(y_,1))
        #求平均 
        accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
        j = 0
        while j<100:
            #gpu_options=tf.GPUOptions(per_process_gpu_memory_fraction=0.333)
            #with tf.Session(config=tf.ConfigProto(gpu_options=gpu_options)) as sess:
            with tf.Session() as sess:
                ckpt = tf.train.get_checkpoint_state(mnist_lenet_backward.MODEL_SAVE_PATH)
                if ckpt and ckpt.model_checkpoint_path:
                    saver.restore(sess, ckpt.model_checkpoint_path)
                    global_step = ckpt.model_checkpoint_path.split('/')[-1].split('-')[-1]
#                     print("global_step:",global_step)
                    xs, ys = mnist.test.next_batch(mnist_lenet_backward.BATCH_SIZE)
                    xs = np.reshape(xs,(mnist_lenet_backward.BATCH_SIZE,
                                      mnist_lenet_forward.IMAGE_SIZE,
                                      mnist_lenet_forward.IMAGE_SIZE,
                                      mnist_lenet_forward.NUM_CHANNELS))
                    accuracy_score = sess.run(accuracy,feed_dict = {x: xs, y_: ys})
                    print("global_step:",global_step)
                    print("After %s training step(s). test accuracy = %f"%(global_step, accuracy_score))
                else:
                    print("No checkpoint file found")
                    return
#                 time.sleep(TEST_INTERVAL_SECS)
                j += 1
def main():
    mnist = input_data.read_data_sets("./MNIST_data/",one_hot = True)
    test(mnist)
    
if __name__ == "__main__":
    main()
        
```
